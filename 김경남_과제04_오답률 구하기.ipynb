```python
# csv파일 데이터 pandas로 읽기 
df = pd.read_csv("python_score.csv",header=None)
df.columns = ['학생1','학생2','학생3','학생4','학생5','학생6','학생7','학생8','학생9','학생10','정답']
df.index =['객1','객2','객3','객4','객5','객6','객7','객8','객9','객10','객11','객12','객13','객14','객15','객16','객17','객18','객19','객20','주1','주2']
df1=df.copy()
df1
```

```python
#학생 답안지: 주관식, 정답 제거
s_answer=df1.iloc[:-2,:-1]
s_answer
```

```python
# 정답
answer=['정답']
answer2 = df1.reindex(columns=answer)
r_answer = answer2[:-2]
r_answer
```

```python
# 오답률 생성, 정답과 비교한 후 블리언 값 컬럼단위로 평균하여 오답률에 대입
s_answer['학생1']= s_answer.학생1==r_answer['정답']
s_answer['학생2']= s_answer.학생2==r_answer['정답']
s_answer['학생3']= s_answer.학생3==r_answer['정답']
s_answer['학생4']= s_answer.학생4==r_answer['정답']
s_answer['학생5']= s_answer.학생5==r_answer['정답']
s_answer['학생6']= s_answer.학생6==r_answer['정답']
s_answer['학생7']= s_answer.학생7==r_answer['정답']
s_answer['학생8']= s_answer.학생8==r_answer['정답']
s_answer['학생9']= s_answer.학생9==r_answer['정답']
s_answer['학생10']= s_answer.학생10==r_answer['정답']
s_answer['오답률(%)']= (1-s_answer.mean(axis=1))*100
s_answer['No.']=np.arange(20)+1
s_answer
```
